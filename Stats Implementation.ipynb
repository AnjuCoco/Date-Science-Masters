```python
# to check whether the plot/data has a normal distribution or not
# check by plotting Q-Q (quantile-quantile plot) as indicated below 
```

```python
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab
```

```python
def plot_data(sample):
    plt.figure(figsize = (10,5))
    plt.subplot(1,2,1)
    sns.histplot(sample, kde = True)
    plt.subplot(1,2,2)
    stat.probplot(sample, dist = 'norm', plot = pylab)
```

```python
var = np.random.lognormal(3,0.5, 5000)
```

```python
sns.histplot(var, kde = True)
```

```python
plot_data(var)
```

```python
# histogram shows that it is a left skewed data 
# and the probability plot shows that the datapoints(in blue colour) are not aligned with the straight line
# which means it is not a normal distribution
```

```python
# let us do the log transformation and convert log normal distribution into normal distn
# by taking log value of all datapoints in log normal distn
var1 = np.log(var)
```

```python
plot_data(var1)
```

```python
# we can see now that the distribution is a normal distribution
# the probability graph shows that most of the datapoints are aligned with the straight line. therefore normal distn
```

```python
sns.distplot(var1)
```

```python
# central limit theorem
```

```python
from statistics import mode
```

```python
# generate random 50 nos. between 30 and 100
population = np.random.randint(30,500, 150)
```

```python
population
```

```python
len(population)
```

```python
print(f"mean of population {np.mean(population)}")
print(f"median of population {np.median(population)}")
print(f"mode of population {mode(population)}")
```

```python
sns.histplot(population, kde = True)
```

```python
# it shows that the population plot is not normal distribution
# to convert this plot into normal distn plot, we will take samples and 
# calculate the mean of each sample and then plot the means 
# applying central limit theorem 
```

```python
sample = np.random.choice(population, 30)
```

```python
print(f"mean of sample {np.mean(sample)}")
print(f"median of sample {np.median(sample)}")
print(f"mode of sample {mode(sample)}")
```

```python
# generate 4 samples
s1 = np.random.choice(population, 40)
s2 = np.random.choice(population, 40)
s3 = np.random.choice(population, 40)
s4 = np.random.choice(population, 40)
```

```python
# calculate mean of all samples and then plot
```

```python
mean_of_samples = []
samples = [s1, s2, s3, s4]
for i in samples:
    mean_of_samples.append(np.mean(i))
```

```python
sns.histplot(mean_of_samples, kde = True)
```

```python
# data points are not enough to get the normal distribution
```

```python
population = np.random.binomial(10, 0.5, 10000)
# 10 - no. of trials
# 0.5 - probability of success(in binomial, probability of success - 1/2)
# 10000 - no. of data points
```

```python
population
```

```python
sns.distplot(population)
```

```python
# to convert the population distribution into a normal distribution using Central Limit Theorem
```

```python
size_sample = int(len(population)*.3)
```

```python
size_sample
```

```python
mean_of_samples = []
for i in range(1, 1000):
    sample = np.random.choice(population, size=size_sample)
    mean_of_samples.append(np.mean(sample))
```

```python
sns.distplot(mean_of_samples, kde = True)
```
