```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/sunnysavita10/Statistics-With-Python-TheCompleteGuide/main/googleplaystore.csv")
df
```

```python
df_copy = df.copy()
```

```python
# converting "Size" column from Object to float 
df_copy["Size"].info
```

```python
df_copy["Size"].unique()
```

```python
df_copy["Size"] = df_copy["Size"].str.replace("M", "000").str.replace("k","").str.replace("Varies with device",str(np.nan))
```

```python
df_copy["Size"].unique()
```

```python
char_to_remove = ["+", "-", ",", "$"]
cols_to_clean = ["Size", "Installs", "Price"]
```

```python
for char in char_to_remove:
    for col in cols_to_clean:
        df_copy[col] = df_copy[col].str.replace(char, "")
```

```python
df_copy["Size"].unique()
```

```python
df_copy["Size"] = df_copy["Size"].astype("float")
```

```python
for i in df_copy["Size"]:
    if i <= 10:
        df_copy["Size"] = df_copy["Size"].replace(i, i*1000)
```

```python
df_copy["Size"].dtype
```

```python
df_copy.info()
```

```python
df_copy["Price"].unique()
```

```python
df_copy["Price"] = df_copy["Price"].str.replace("Everyone", str(np.nan))
```

```python
df_copy["Price"].unique()
```

```python
df_copy["Price"].dtype
```

```python
df_copy["Price"] = df_copy["Price"].astype("float")
```

```python
df_copy["Price"].dtype
```

```python
df_copy.info()
```

```python
df_copy["Installs"].dtype 
```

```python
df_copy["Installs"] = df_copy["Installs"].astype('int')
```

```python
df_copy["Reviews"] = df_copy["Reviews"].astype("float")
```

```python
# to find out which category has highest installation

df_copy.groupby(["Category"])["Installs"].sum().sort_values(ascending = False).reset_index()
```

```python
# to group more than two columns

df_copy.groupby(["Category", "Type"])["Installs"].sum().sort_values(ascending = False).reset_index()
```

```python
# which categories are free

df_temp = df_copy.groupby(["Category", "Type"])["Installs"].sum().sort_values(ascending = False).reset_index()
```

```python
df_temp
```

```python
df_temp[df_temp.Type == "Free"]
```
