```python
# to find out max. fours scored by which batsman
mask = delivery['batsman_runs']==4
max_fours = delivery[mask]
```

```python
# create a function for fetching the name of the batsman
def batsman(batsman_runs):
    b_name = delivery[delivery['batsman_runs']==batsman_runs]
    return b_name.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False).head(1)
```

```python
# calling function to check who has scored maximum four runs
batsman(4)
```

```python
batsman(6)
# calling function to check who has scored maximum six runs
```

```python
batsman(3)
# calling function to check who has scored maximum three runs
```

```python
# creating a function
def bowling_team(batsman):
    bm = delivery[delivery['batsman']==batsman]
    return bm.groupby('bowling_team').size().sort_values(ascending=False).head(1)
```

```python
# Find out the most destructive death over batsman in history of IPL based on strike rate and
# minimum no. of balls played by the batsman >200
```

```python
mask = delivery['over']>15
death_over = delivery[mask]
```

```python
death_over
```

```python
all_batsman = death_over.groupby('batsman')['batsman_runs'].count()
```

```python
x = all_batsman>200
x
```

```python
batsman_list = all_batsman[x]
```

```python
all_batsman[x].size
```

```python
# to find out the strike rate
```

```python
# find no. of runs and no. of balls played
```

```python
mask = delivery['batsman'].isin(batsman_list)
```

```python
final = delivery[mask]
```

```python
runs = final.groupby('batsman')['batsman_runs'].sum()
runs
```

```python
balls = final.groupby('batsman')['batsman_runs'].count()
```

```python
sr = runs/balls*100
```

```python
sr.sort_values(ascending=False)
```
