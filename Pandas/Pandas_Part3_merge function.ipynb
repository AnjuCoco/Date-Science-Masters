```python
# merge function
# Problem statement - List of Orange cap holders for each season (Orange cap holder - names of batsman who scored maximum runs in each season)
```

```python
# We can't get the required data from any one dataset. Hence, we will have to merge these two datasets
# To merge, we need one common column, it is id - match id in deliveries and id in matches
```

```python
deliveries.merge(matches, left_on=	'match_id', right_on='id').head(3)
```

```python
# seting up the index
```

```python
deliveries.merge(matches, left_on='match_id', right_on='id').set_index('match_id')
```

```python
# making the index change permanent and creating a new dataset after merger
new = deliveries.merge(matches, left_on='match_id', right_on='id').set_index('match_id')
```

```python
new.head()
```

```python
new.info()
```

```python
new.groupby(['season','batsman'])['batsman_runs'].sum()
# gives season wise each batsmans total no. of runs scored
```

```python
# converting season index into a column using reset_index
```

```python
new.groupby(['season','batsman'])['batsman_runs'].sum().reset_index()
```

```python
# storing new dataframe into a variable
final = new.groupby(['season','batsman'])['batsman_runs'].sum().reset_index()
```

```python
final.head(10)
```

```python
final.sort_values(['season','batsman_runs'], ascending = False).head()
```

```python
# keeping only the one who scored the highest using drop_duplicates
```

```python
final.sort_values(['season','batsman_runs'], ascending = False).drop_duplicates('season',keep='first')
```

```python
# now I want an output as season and batsman only
```

```python
orangecapwinner = final.sort_values(['season', 'batsman_runs'],ascending =False).drop_duplicates('season',keep='first')
```
